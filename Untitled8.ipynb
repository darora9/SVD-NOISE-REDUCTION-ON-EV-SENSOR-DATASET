```python
data = pd.read_csv('/content/drive/MyDrive/battery data/nasa dataset/B05_discharge_soh.csv')

# Display the first few rows of the dataset
data.head()
```

```python
data = data.drop(columns=['time','capacity','cycle','SOH'])
```

```python
# Display the first few rows of the dataset
data
```

```python
# Generate random row indices
np.random.seed(50)
random_indices = np.random.choice(data.shape[0], size=10000, replace=False)

# Select rows using these indices
selected_data = data.iloc[random_indices]

selected_data
```

```python
scaler = StandardScaler()
standardized_data = scaler.fit_transform(selected_data)
df_standardized_data= pd.DataFrame(standardized_data)
df_standardized_data
```

```python
m= df_standardized_data.shape[0]
m
```

```python
n = df_standardized_data.shape[1]
n
```

```python
# Perform SVD
U, Sigma, VT = np.linalg.svd(df_standardized_data, full_matrices=True)

S = np.zeros((df_standardized_data.shape[0], df_standardized_data.shape[1]))
np.fill_diagonal(S, Sigma)
```

```python
df_U= pd.DataFrame(U)
df_U
```

```python
df_S= pd.DataFrame(S)
df_S
```

```python
df_VT= pd.DataFrame(VT)
df_VT
```

```python
# Calculate the explained variance
explained_variance = (Sigma ** 2) / np.sum(Sigma ** 2)
cumulative_explained_variance = np.cumsum(explained_variance)

# Plot the cumulative explained variance to find the elbow point
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o')
plt.xlabel('Number of Components(k)')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance vs. Number of Components(k)')
plt.grid(True)
plt.show()
```

```python
k = 3 #using elbow method

dropped = n-k

U_reduced = U[:, :k]
df_u_reduced= pd.DataFrame(U_reduced)

S_reduced = np.zeros((k, VT.shape[1] - dropped))
np.fill_diagonal(S_reduced, Sigma[:k])
df_s_reduced= pd.DataFrame(S_reduced)

VT_reduced = VT[:k, :]
df_VT_reduced= pd.DataFrame(VT_reduced)
```

```python
reduced_data = np.dot(df_u_reduced, np.dot(df_s_reduced, df_VT_reduced))

df_reduced_data= pd.DataFrame(reduced_data)
df_reduced_data
```

```python
# Explained variance
explained_variance_retained = cumulative_explained_variance[k-1] * 100  # percentage
print(f"\nExplained variance retained with k={k}: {explained_variance_retained:.2f}%")
```
